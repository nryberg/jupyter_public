```python
import requests
from bs4 import BeautifulSoup
from IPython.display import display_html
import pandas as pd
import csv
import base64
from IPython.display import HTML
import reprlib
```

```python
r = reprlib.Repr()
r.maxlist = 10       # max elements displayed for lists
r.maxstring = 50    # max characters displayed for strings
```

## Code for Export

```python
def create_download_link( df, title = "Download CSV file", filename = "data.csv"):
    csv = df.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

# Get the Data

```python
url = "https://www.world-airport-codes.com/us-top-40-airports.html"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
page = requests.get(url, headers=headers)
```

```python
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find("table", {"class": "wikitable sortable"})
```

```python
print(r.repr(results.prettify()))
```

```python
display_html(results.prettify(), raw=True)
```

```python
dfs = pd.read_html(results.prettify())
df = dfs[0]
df.head(3)
```

```python
df.rename(columns={'IATA  Code':'Code','Major city served':'Major City'}, inplace = True)
```

```python
df.columns
```

## unpivot

```python
unpivot = pd.melt(df, id_vars=['Code','Major City'], value_vars=[ '2018','2017', '2016'], value_name='travellers', var_name='year')
```

```python
unpivot.head(5)
```

# Export the file

```python
create_download_link(unpivot)
```
